## 5. Converting Alex data to pandas dataframe

#### Now we are ready to clean the json and structure it into a pandas dataframe

#### From the Foursquare lab in the previous module, we know that all the information is in the items key. Before we proceed, let's borrow the get_category_type function from the Foursquare lab.

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

#### Cleaning and filtering columns

```python
alex = results['response']['groups'][0]['items']
    
df_alex = json_normalize(alex) # flatten JSON

# filter columns
filtered_columns = ['venue.id','venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
df_alex =df_alex.loc[:, filtered_columns]

# filter the category for each row
df_alex['venue.categories'] = df_alex.apply(get_category_type, axis=1)

# clean columns
df_alex.columns = [col.split(".")[-1] for col in df_alex.columns]

df_alex
```

#### using folium (plotting library) to mapping the data

```python
ALEX_MAP = folium.Map(location=[latitude, longitude],tiles= "Mapbox Bright", zoom_start=10) # generate map centred around Alex
# add a red circle marker to represent ALex
folium.CircleMarker([latitude, longitude],
    radius=10,
    color='red',
    popup='Alex',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(ALEX_MAP)

# add the best 20 places in Alexandria
for lat, lng, label in zip(df_alex.lat, df_alex.lng, df_alex.name):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(ALEX_MAP)
 # display map
ALEX_MAP
```
